### Utilisation de la méthode map pour remplacer les valeurs dans une colonne en passant un dictionnaire de la forme {ancienne_valeur:nouvelle_valeur}

```python
#la variable International plan possède les modalités No et Yes. Remplaçons No par False et Yes par true
d={'No':False,'Yes':True}
df['International plan']=df['International plan'].map(d)
df
```

```python
# Autre méthode
df=df.replace({'Voice mal plan':d})
df[:1]
```

### Tableaux croisés

```python
#Supposons que nous voulions connaitre les effectifs ou fréquences des modalités de la variable churn selon la variable International plan
#effectifs
pd.crosstab(df['Churn'],df['International plan'])
```

```python
#fréquences
pd.crosstab(df['Churn'],df['International plan'],normalize=True)
```

### Tableau croisé dynamique

```python
# affichons le nbre total d'abonnés qu'effectuent des appels le jours, le soir et la nuit par indicatif régional
df.pivot_table(['Total day calls','Total eve calls','Total night calls'],['Area code'],aggfunc='sum')
```
