### Utilisation de la m√©thode map pour remplacer les valeurs dans une colonne en passant un dictionnaire de la forme {ancienne_valeur:nouvelle_valeur}

```python
#la variable International plan poss√®de les modalit√©s No et Yes. Rempla√ßons No par False et Yes par true
d={'No':False,'Yes':True}
df['International plan']=df['International plan'].map(d)
df
```

```python
# Autre m√©thode
df=df.replace({'Voice mal plan':d})
df[:1]
```

### Tableaux crois√©s

```python
#Supposons que nous voulions connaitre les effectifs ou fr√©quences des modalit√©s de la variable churn selon la variable International plan
#effectifs
pd.crosstab(df['Churn'],df['International plan'])
```

```python
#fr√©quences
pd.crosstab(df['Churn'],df['International plan'],normalize=True)
```

### Tableau crois√© dynamique

```python
# affichons le nbre total d'abonn√©s qu'effectuent des appels le jours, le soir et la nuit par indicatif r√©gional
df.pivot_table(['Total day calls','Total eve calls','Total night calls'],['Area code'],aggfunc='sum')
```
